```python
df = df.drop(
                [
                    "order_approved_at",
                    "order_delivered_carrier_date",
                    "order_delivered_customer_date",
                    "order_estimated_delivery_date",
                    "order_purchase_timestamp",
                ],axis=1)
df.info()
```

```python
import numpy as np

# 1. Fill specific columns (Avoid inplace=True for better reliability)


# 1. Fill specific product dimensions with Median
cols_to_median = ["product_weight_g", "product_length_cm", "product_height_cm", "product_width_cm"]
for col in cols_to_median:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

# 2. Handle the text column
if "review_comment_message" in df.columns:
    df["review_comment_message"] = df["review_comment_message"].fillna("No review")

# 3. Select Numeric Types ONLY
# This is crucial for Linear Regression
df = df.select_dtypes(include=[np.number])

# 4. THE SAFETY NET: Fill ANY remaining NaNs in numeric columns
# We use the median of the current numeric dataframe
df = df.fillna(df.median())

# 5. Drop non-predictive IDs
# FIX: Removed axis=1 because we are using the 'columns' parameter
cols_to_drop = ["customer_zip_code_prefix", "order_item_id"]
df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

# --- VERIFICATION STEPS (Great for Notebooks) ---
print(f"Final shape of data: {df.shape}")
print(f"Remaining NaN values: {df.isna().sum().sum()}")

# Show the first few rows to ensure it looks right
df.head()
```
